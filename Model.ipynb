```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, AdaBoostRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
algorithms = {}
```

```python
# 1. Random Forest (Usually excellent for tabular data)
print("\n1. Setting up Random Forest...")
rf_reg = MultiOutputRegressor(RandomForestRegressor(
    n_estimators=500,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
))
algorithms['Random Forest'] = rf_reg
```

```python
# 2. Extra Trees (Often better than Random Forest)
print("2. Setting up Extra Trees...")
et_reg = MultiOutputRegressor(ExtraTreesRegressor(
    n_estimators=500,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
))
algorithms['Extra Trees'] = et_reg
```

```python
# 3. Gradient Boosting (Sequential boosting)
print("3. Setting up Gradient Boosting...")
gb_reg = MultiOutputRegressor(GradientBoostingRegressor(
    n_estimators=300,
    max_depth=8,
    learning_rate=0.1,
    subsample=0.8,
    random_state=42
))
algorithms['Gradient Boosting'] = gb_reg
```

```python
# 5. Neural Network (MLP) with Pipeline
print("5. Setting up Neural Network...")
mlp_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', MultiOutputRegressor(MLPRegressor(
        hidden_layer_sizes=(200, 100, 50),
        max_iter=1000,
        random_state=42,
        early_stopping=True,
        validation_fraction=0.1,
        alpha=0.001
    )))
])
algorithms['Neural Network'] = mlp_pipeline
```

```python
# 6. Support Vector Regression with Pipeline
print("6. Setting up SVR...")
svr_pipeline = Pipeline([
    ('scaler', StandardScaler()),  
    ('regressor', MultiOutputRegressor(SVR(
        kernel='rbf',
        C=100,
        gamma='scale',
        epsilon=0.1
    )))
])
algorithms['SVR'] = svr_pipeline
```

```python
# 7. Ridge Regression (Regularized Linear)
print("7. Setting up Ridge Regression...")
ridge_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', MultiOutputRegressor(Ridge(alpha=1.0)))
])
algorithms['Ridge Regression'] = ridge_pipeline
```

```python
# 8. ElasticNet (L1 + L2 regularization)
print("8. Setting up ElasticNet...")
elastic_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', MultiOutputRegressor(ElasticNet(alpha=0.1, l1_ratio=0.5)))
])
algorithms['ElasticNet'] = elastic_pipeline
```

```python
results = {}
for name, algorithm in algorithms.items():
    print(f"\nTraining {name}...")
    
    try:
        # Train the algorithm
        algorithm.fit(X_train, y_train)
        
        # Make predictions
        y_pred = algorithm.predict(X_test)
        
        # Calculate metrics for each target variable
        r2_scores = []
        mse_scores = []
        mae_scores = []
        
        for i, target in enumerate(target_columns):
            r2 = r2_score(y_test.iloc[:, i], y_pred[:, i])
            mse = mean_squared_error(y_test.iloc[:, i], y_pred[:, i])
            mae = mean_absolute_error(y_test.iloc[:, i], y_pred[:, i])
            
            r2_scores.append(r2)
            mse_scores.append(mse)
            mae_scores.append(mae)
        
        # Calculate overall R² score
        overall_r2 = r2_score(y_test, y_pred)
        accuracy_percentage = overall_r2 * 100
        
        results[name] = {
            'Overall_R2': overall_r2,
            'Accuracy_%': accuracy_percentage,
            'Individual_R2': dict(zip(target_columns, r2_scores)),
            'MSE': dict(zip(target_columns, mse_scores)),
            'MAE': dict(zip(target_columns, mae_scores)),
            'Status': 'Success'
        }
        
    except Exception as e:
        print(f"Error training {name}: {str(e)}")
        results[name] = {'Status': f'Failed: {str(e)}'}
```

```python
print("\n" + "="*80)
print("ALGORITHM PERFORMANCE COMPARISON")
print("="*80)

successful_results = {k: v for k, v in results.items() if v.get('Status') == 'Success'}
sorted_results = sorted(successful_results.items(), key=lambda x: x[1]['Accuracy_%'], reverse=True)

for name, metrics in sorted_results:
    print(f"\n{name}:")
    print(f"  Overall Accuracy: {metrics['Accuracy_%']:.2f}%")
    print(f"  Overall R²: {metrics['Overall_R2']:.4f}")
    
    for target in target_columns:
        print(f"  {target} - R²: {metrics['Individual_R2'][target]:.4f}, "
              f"MSE: {metrics['MSE'][target]:.4f}, MAE: {metrics['MAE'][target]:.4f}")

failed_results = {k: v for k, v in results.items() if v.get('Status') != 'Success'}
if failed_results:
    print(f"\n  Failed algorithms: {list(failed_results.keys())}")

print("\n" + "="*80)
print("RECOMMENDATIONS:")
print("="*80)

if sorted_results:
    best_algorithm = sorted_results[0]
    print(f" BEST ALGORITHM: {best_algorithm[0]} with {best_algorithm[1]['Accuracy_%']:.2f}% accuracy")
    
    algorithms_above_90 = [name for name, metrics in successful_results.items() if metrics['Accuracy_%'] >= 90]
    if algorithms_above_90:
        print(f" Algorithms with >90% accuracy: {', '.join(algorithms_above_90)}")
    else:
        print("  No algorithm achieved >90% accuracy with default parameters.")
        print(" Try hyperparameter tuning on the best performing algorithm:")
        print(f"   Best: {best_algorithm[0]} with {best_algorithm[1]['Accuracy_%']:.2f}%")

    print(f"\n Total features used: {X_train.shape[1]}")
    print(f" Sample size: {len(final_df)}")
    
    best_model_name = best_algorithm[0]
    best_model = algorithms[best_model_name]
    print(f"\n Best model stored as: {best_model_name}")

else:
    print(" All algorithms failed. Check your data preprocessing.")
```
