```python
from datetime import datetime

def calculate_doy(row):
    date = datetime(int(row["YEAR"]), int(row["MO"]), int(row["DY"]))
    return date.timetuple().tm_yday

# Apply to the base DataFrame (after merging df3, df4, df5)
df_base = df3.merge(df4, on=["YEAR", "MO", "DY"]) \
             .merge(df5, on=["YEAR", "MO", "DY"])

# Add DOY column
df_base["DOY"] = df_base.apply(calculate_doy, axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train_rain, X_test_rain, y_train_rain, y_test_rain = train_test_split(X_rainfall, y_rainfall, test_size=0.2, random_state=42)

X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(X_temperature, y_temperature, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,accuracy_score,r2_score

# Rainfall
model_rain = LinearRegression()
model_rain.fit(X_train_rain, y_train_rain)
y_pred_rain = model_rain.predict(X_test_rain)
rain_mse = mean_squared_error(y_test_rain, y_pred_rain)

# Temperature
model_temp = LinearRegression()
model_temp.fit(X_train_temp, y_train_temp)
y_pred_temp = model_temp.predict(X_test_temp)
temp_mse = mean_squared_error(y_test_temp, y_pred_temp)
r2 = r2_score(y_test_rain, y_pred_rain)

print("Rainfall MSE:", rain_mse)
print("Temperature MSE:", temp_mse)
print("Rainfall R² Score:", r2)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, AdaBoostRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
algorithms = {}
```

```python
print("\n1. Setting up Random Forest...")
rf_reg = MultiOutputRegressor(RandomForestRegressor(
    n_estimators=500,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
))
algorithms['Random Forest'] = rf_reg
```

```python
print("2. Setting up Extra Trees...")
et_reg = MultiOutputRegressor(ExtraTreesRegressor(
    n_estimators=500,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
))
algorithms['Extra Trees'] = et_reg
```

```python
print("3. Setting up Gradient Boosting...")
gb_reg = MultiOutputRegressor(GradientBoostingRegressor(
    n_estimators=300,
    max_depth=8,
    learning_rate=0.1,
    subsample=0.8,
    random_state=42
))
algorithms['Gradient Boosting'] = gb_reg
```

```python
print("5. Setting up Neural Network...")
mlp_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', MultiOutputRegressor(MLPRegressor(
        hidden_layer_sizes=(200, 100, 50),
        max_iter=1000,
        random_state=42,
        early_stopping=True,
        validation_fraction=0.1,
        alpha=0.001
    )))
])
algorithms['Neural Network'] = mlp_pipeline
```

```python
print("6. Setting up SVR...")
svr_pipeline = Pipeline([
    ('scaler', StandardScaler()),  
    ('regressor', MultiOutputRegressor(SVR(
        kernel='rbf',
        C=100,
        gamma='scale',
        epsilon=0.1
    )))
])
algorithms['SVR'] = svr_pipeline
```

```python
print("7. Setting up Ridge Regression...")
ridge_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', MultiOutputRegressor(Ridge(alpha=1.0)))
])
algorithms['Ridge Regression'] = ridge_pipeline
```

```python
print("8. Setting up ElasticNet...")
elastic_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', MultiOutputRegressor(ElasticNet(alpha=0.1, l1_ratio=0.5)))
])
algorithms['ElasticNet'] = elastic_pipeline
```

```python
y_train_rain = y_train_rain.to_frame()
y_test_rain = y_test_rain.to_frame()
y_train_temp = y_train_temp.to_frame()
y_test_temp = y_test_temp.to_frame()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

results = {}
for name, algorithm in algorithms.items():
    print(f"\nTraining {name}...")

    try:
        algorithm.fit(X_train_rain, y_train_rain)
        y_pred_rain = algorithm.predict(X_test_rain)

        r2_rain = r2_score(y_test_rain, y_pred_rain)
        mse_rain = mean_squared_error(y_test_rain, y_pred_rain)
        mae_rain = mean_absolute_error(y_test_rain, y_pred_rain)

        algorithm.fit(X_train_temp, y_train_temp)
        y_pred_temp = algorithm.predict(X_test_temp)

        r2_temp = r2_score(y_test_temp, y_pred_temp)
        mse_temp = mean_squared_error(y_test_temp, y_pred_temp)
        mae_temp = mean_absolute_error(y_test_temp, y_pred_temp)

        avg_r2 = (r2_rain + r2_temp) / 2
        avg_accuracy = avg_r2 * 100

        results[name] = {
            'Status': 'Success',
            'Accuracy_%': avg_accuracy,
            'Overall_R2': avg_r2,
            'Individual_R2': {'Rain': r2_rain, 'Temp': r2_temp},
            'MSE': {'Rain': mse_rain, 'Temp': mse_temp},
            'MAE': {'Rain': mae_rain, 'Temp': mae_temp},
            'Sample_Size': len(y_test_rain) + len(y_test_temp)  # Optional
        }

    except Exception as e:
        print(f"Error training {name}: {str(e)}")
        results[name] = {'Status': f'Failed: {str(e)}'}
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

print("\n" + "="*80)
print("ALGORITHM PERFORMANCE COMPARISON")
print("="*80)

successful_results = {
    k: v for k, v in results.items()
    if v.get('Status') == 'Success' and 'Accuracy_%' in v and 'Overall_R2' in v
}

sorted_results = sorted(successful_results.items(), key=lambda x: x[1].get('Accuracy_%', 0), reverse=True)

for name, metrics in sorted_results:
    print(f"\n{name}:")
    print(f"  Overall Accuracy: {metrics.get('Accuracy_%', float('nan')):.2f}%")
    print(f"  Overall R²: {metrics.get('Overall_R2', float('nan')):.4f}")

    for target in ['Rain', 'Temp']:
        r2 = metrics.get('Individual_R2', {}).get(target, float('nan'))
        mse = metrics.get('MSE', {}).get(target, float('nan'))
        mae = metrics.get('MAE', {}).get(target, float('nan'))
        rmse = metrics.get('RMSE', {}).get(target, np.sqrt(mse) if not np.isnan(mse) else float('nan'))

        print(f"  {target} - R²: {r2:.4f}, MSE: {mse:.4f}, MAE: {mae:.4f}, RMSE: {rmse:.4f}")

failed_results = {k: v for k, v in results.items() if v.get('Status') != 'Success'}
if failed_results:
    print(f"\nFailed algorithms: {list(failed_results.keys())}")


print("\n" + "="*80)
print("RECOMMENDATIONS:")
print("="*80)

if sorted_results:
    best_algorithm = sorted_results[0]
    best_name = best_algorithm[0]
    best_metrics = best_algorithm[1]

    print(f" BEST ALGORITHM: {best_name} with {best_metrics.get('Accuracy_%', float('nan')):.2f}% accuracy")

    algorithms_above_90 = [
        name for name, metrics in successful_results.items()
        if metrics.get('Accuracy_%', 0) >= 90
    ]

    if algorithms_above_90:
        print(f" Algorithms with >90% accuracy: {', '.join(algorithms_above_90)}")
    else:
        print("  No algorithm achieved >90% accuracy with default parameters.")
        print(" Try hyperparameter tuning on the best performing algorithm:")
        print(f"   Best: {best_name} with {best_metrics.get('Accuracy_%', 0):.2f}%")

    if 'X_train_rain' in globals():
        print(f"\n Features for Rain prediction: {X_train_rain.shape[1]}")
    if 'X_train_temp' in globals():
        print(f" Features for Temp prediction: {X_train_temp.shape[1]}")

    print(f" Sample size: {best_metrics.get('Sample_Size', 'N/A')}")

    best_model = algorithms.get(best_name, None)
    if best_model:
        print(f"\n Best model stored as: {best_name}")
    else:
        print(" Warning: Best model object not found in algorithms dict!")

else:
    print(" All algorithms failed. Check your data preprocessing or model compatibility.")
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np


scaler_rain = StandardScaler()
X_train_rain_scaled = scaler_rain.fit_transform(X_train_rain)
X_test_rain_scaled = scaler_rain.transform(X_test_rain)

scaler_temp = StandardScaler()
X_train_temp_scaled = scaler_temp.fit_transform(X_train_temp)
X_test_temp_scaled = scaler_temp.transform(X_test_temp)


y_train_rain = y_train_rain.values.ravel()
y_train_temp = y_train_temp.values.ravel()


param_grid = {
    'hidden_layer_sizes': [(128,), (256,), (128, 64), (256, 128), (128, 128, 64)],
    'activation': ['relu', 'tanh'],
    'alpha': [0.0001, 0.001, 0.005],
    'learning_rate': ['constant', 'adaptive'],
    'learning_rate_init': [0.001, 0.01, 0.05],
    'max_iter': [1000]
}


print("\n🔧 Tuning Neural Network for Rain prediction...")
nn_rain = MLPRegressor(random_state=42, early_stopping=True, n_iter_no_change=10)
rand_rain = RandomizedSearchCV(
    nn_rain,
    param_distributions=param_grid,
    n_iter=30,
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)
rand_rain.fit(X_train_rain_scaled, y_train_rain)
best_nn_rain = rand_rain.best_estimator_
print("Best parameters for Rain:", rand_rain.best_params_)

y_pred_rain = best_nn_rain.predict(X_test_rain_scaled)
r2_rain = r2_score(y_test_rain, y_pred_rain)
mse_rain = mean_squared_error(y_test_rain, y_pred_rain)
mae_rain = mean_absolute_error(y_test_rain, y_pred_rain)
rmse_rain = np.sqrt(mse_rain)
print(f"Rain R²: {r2_rain:.4f}, MSE: {mse_rain:.4f}, MAE: {mae_rain:.4f}, RMSE: {rmse_rain:.4f}")


print("\n🔧 Tuning Neural Network for Temp prediction...")
nn_temp = MLPRegressor(random_state=42, early_stopping=True, n_iter_no_change=10)
rand_temp = RandomizedSearchCV(
    nn_temp,
    param_distributions=param_grid,
    n_iter=30,
    cv=3,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)
rand_temp.fit(X_train_temp_scaled, y_train_temp)
best_nn_temp = rand_temp.best_estimator_
print("Best parameters for Temp:", rand_temp.best_params_)

y_pred_temp = best_nn_temp.predict(X_test_temp_scaled)
r2_temp = r2_score(y_test_temp, y_pred_temp)
mse_temp = mean_squared_error(y_test_temp, y_pred_temp)
mae_temp = mean_absolute_error(y_test_temp, y_pred_temp)
rmse_temp = np.sqrt(mse_temp)
print(f"Temp R²: {r2_temp:.4f}, MSE: {mse_temp:.4f}, MAE: {mae_temp:.4f}, RMSE: {rmse_temp:.4f}")


overall_r2 = (r2_rain + r2_temp) / 2
accuracy_percent = overall_r2 * 100
print("\n" + "="*60)
print(f"NEURAL NETWORK ACCURACY: {accuracy_percent:.2f}%")
print("="*60)

if accuracy_percent >= 90:
    print("Achieved >90% accuracy!")
else:
    print("Accuracy below 90%. Consider further tuning:")
    print("- Add more data or features")
    print("- Try deeper networks or ensemble stacking")
    print("- Use feature selection or PCA")
```
